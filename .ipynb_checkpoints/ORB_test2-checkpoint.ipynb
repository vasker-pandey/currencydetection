```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Function to extract ORB keypoints and return a 2x2 matrix of keypoint coordinates
def extract_orb_keypoints(image_path):
    # Read the image in grayscale
    image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
    
    if image is None:
        raise ValueError(f"Image at path '{image_path}' could not be loaded.")
    
    # Initialize ORB detector
    orb = cv2.ORB_create()

    # Detect keypoints and compute descriptors (we don't need descriptors here)
    kp, des = orb.detectAndCompute(image, None)

    # Check if keypoints are available
    if not kp:
        raise ValueError(f"No keypoints found in the image at '{image_path}'.")

    # Extract the first 4 keypoints' coordinates (x, y) into a list
    kp_coordinates = np.array([kp[i].pt for i in range(min(len(kp), 4))], dtype=np.float32)
    
    # If there are fewer than 4 keypoints, pad the array with zeros
    if len(kp_coordinates) < 4:
        kp_coordinates = np.pad(kp_coordinates, ((0, 4 - len(kp_coordinates)), (0, 0)), mode='constant')

    # Reshape the keypoints coordinates to be a 2x2 matrix
    kp_matrix = kp_coordinates.reshape(2, 2, 2)

    return kp_matrix, kp

# Function to plot keypoints on the image
def plot_keypoints(image_path, keypoints):
    # Read the image in grayscale
    image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

    # Draw keypoints on the image
    image_with_keypoints = cv2.drawKeypoints(image, keypoints, None, color=(0,255,0))

    # Plot the image with keypoints
    plt.imshow(image_with_keypoints, cmap='gray')
    plt.title("Keypoints Detected by ORB")
    plt.show()

# Example usage
img1_path = '1.jpg'  # Path to first image
img2_path = '2.jpg'  # Path to second image

try:
    # Get ORB keypoint matrices for both images
    matrix1, keypoints1 = extract_orb_keypoints(img1_path)
    matrix2, keypoints2 = extract_orb_keypoints(img2_path)

    # Display the keypoint matrices
    print("2x2 Keypoint Matrix for Image 1:")
    print(matrix1)

    print("\n2x2 Keypoint Matrix for Image 2:")
    print(matrix2)

    # Plot keypoints for both images
    print("\nKeypoints on Image 1:")
    plot_keypoints(img1_path, keypoints1)

    print("Keypoints on Image 2:")
    plot_keypoints(img2_path, keypoints2)

except ValueError as e:
    print(e)
```
